```python
import pandas as pd

df = pd.read_csv('data/review_extraction_results.csv')
df.head()
```

```python
from rich.pretty import pprint

df[["External Knowledge Data Type", "Source of External Knowledge", "Means of Augmentation", "Phase of Knowledge Augmentation"]].head()
```

```python
columns = ["External Knowledge Data Type", "Source of External Knowledge", "Means of Augmentation", "Phase of Knowledge Augmentation"]

for column in columns:
    pprint(column + ":\n")
    values = df[column].str.split(";")
    flattened_values = []
    for idx, sublist in enumerate(values):
        if isinstance(sublist, list):
            for value in sublist:
                flattened_values.append(value.strip())
        else:
            pprint(df.iloc[idx])
            continue
    value_counts = pd.Series(flattened_values).value_counts().sort_values(ascending=False)
    pprint(value_counts)
```
